```python
date= []
time = []
latitude= []
longitude = []
region_name= []
for i in range(20):
    time.append(soup.select("td.tabev6 > b > a")[i].get_text())
    
time

from dateutil.parser import parse

datetimes = [ parse(time.get_text()) for time in soup.select("td.tabev6 > b > a")[:20] ]

times = [ datetime.time().strftime("%H:%M:%S") for datetime in datetimes ]
dates = [ datetime.date().strftime("%Y-%m-%d") for datetime in datetimes ]

print(times, dates)

datetimes = []
for time in soup.select("td.tabev6 > b > a")[:20]:
    datetimes.append({
        "date": parse(time.get_text()).date().strftime("%Y-%m-%d"),
        "time": parse(time.get_text()).time().strftime("%H:%M:%S")
    })

print(datetimes)
```

```python
df =pd.DataFrame(datetimes)
df
```

```python
data = []
for row in soup.select("#tbody > tr")[:20]:
    datetime = row.select("td.tabev6 > b > a")[0]
    longitude = row.select("td.tabev1")[0]
    data.append({
        "date": parse(datetime.get_text()).date().strftime("%Y-%m-%d"),
        "time": parse(datetime.get_text()).time().strftime("%H:%M:%S"),
        "longitude": longitude.text
    })

print(data)
```
