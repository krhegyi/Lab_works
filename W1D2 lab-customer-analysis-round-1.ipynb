```python
df1.columns= df1.columns.str.lower()
df2.columns= df2.columns.str.lower()
df3.columns= df3.columns.str.lower()
```

```python
df1.columns =['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']
df2.columns =['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'total claim amount', 'vehicle class']
```

```python
df1 = df1[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'vehicle class', 'total claim amount']]
df2 = df2[['customer', 'state', 'gender', 'education', 'customer lifetime value', 'income', 'monthly premium auto', 'number of open complaints', 'policy type', 'total claim amount', 'vehicle class']]
df3 = df3[['customer', 'state', 'gender', 'education', 'customer lifetime value','income', 'monthly premium auto', 'number of open complaints', 'policy type', 'total claim amount', 'vehicle class']]
```

```python
#solution for the customer lifetime value: defin the columns, and loop: if % at the end, then remove that and /100, if else: turn to number
def magic_function(x):
    x = str(x)
    if x.endswith("%"):
        return float(x[:-1])/100
    else:
        return float(x)
```

```python
data["customer lifetime value"].apply(magic_function)
```
